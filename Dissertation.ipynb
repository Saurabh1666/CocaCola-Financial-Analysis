```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy.optimize as sco

# Example asset returns data
data = {
    'Asset_A': [0.1, 0.12, 0.14, 0.10, 0.13],
    'Asset_B': [0.05, 0.07, 0.09, 0.04, 0.06],
    'Asset_C': [0.15, 0.16, 0.14, 0.17, 0.18]
}

# Convert the data to a DataFrame
df = pd.DataFrame(data)

# Calculate expected returns and covariance matrix
returns = df.mean()
cov_matrix = df.cov()

print("Expected Returns:\n", returns)
print("\nCovariance Matrix:\n", cov_matrix)

def portfolio_performance(weights, returns, cov_matrix, risk_free_rate=0.01):
    portfolio_return = np.dot(weights, returns)
    portfolio_volatility = np.sqrt(np.dot(weights.T, np.dot(cov_matrix, weights)))
    sharpe_ratio = (portfolio_return - risk_free_rate) / portfolio_volatility
    return portfolio_return, portfolio_volatility, sharpe_ratio

def negative_sharpe_ratio(weights, returns, cov_matrix, risk_free_rate=0.01):
    return -portfolio_performance(weights, returns, cov_matrix, risk_free_rate)[2]

constraints = {'type': 'eq', 'fun': lambda weights: np.sum(weights) - 1}
bounds = tuple((0, 1) for _ in range(len(returns)))

# Initial guess (equal weights)
initial_guess = np.array([1/len(returns)] * len(returns))

# Perform the optimization
optimized_result = sco.minimize(negative_sharpe_ratio, initial_guess, args=(returns, cov_matrix),
                                method='SLSQP', bounds=bounds, constraints=constraints)

# Extract the optimal weights
optimal_weights = optimized_result.x
print("Optimal Weights:\n", optimal_weights)

optimal_return, optimal_volatility, optimal_sharpe = portfolio_performance(optimal_weights, returns, cov_matrix)
print("\nOptimal Portfolio Performance:")
print("Return:", optimal_return)
print("Volatility:", optimal_volatility)
print("Sharpe Ratio:", optimal_sharpe)

def plot_efficient_frontier(returns, cov_matrix, risk_free_rate=0.01, num_portfolios=10000):
    results = np.zeros((3, num_portfolios))
    weights_record = []

    for i in range(num_portfolios):
        weights = np.random.random(len(returns))
        weights /= np.sum(weights)
        weights_record.append(weights)
        portfolio_return, portfolio_volatility, _ = portfolio_performance(weights, returns, cov_matrix, risk_free_rate)
        results[0,i] = portfolio_return
        results[1,i] = portfolio_volatility
        results[2,i] = (portfolio_return - risk_free_rate) / portfolio_volatility

    plt.figure(figsize=(10, 6))
    plt.scatter(results[1,:], results[0,:], c=results[2,:], cmap='viridis', marker='o')
    plt.colorbar(label='Sharpe Ratio')
    plt.scatter(optimal_volatility, optimal_return, color='red', marker='*', s=100)  # Mark the optimal portfolio
    plt.title('Efficient Frontier')
    plt.xlabel('Volatility (Risk)')
    plt.ylabel('Return')
    plt.show()

plot_efficient_frontier(returns, cov_matrix)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
import statsmodels.api as sm

# Load the Cleaned Data
df = pd.read_excel('/Users/saurabh/Downloads/Cleaned_Dissertation_Dataset.xlsx')

# Step 1: Correlation Analysis (to identify potential relationships)
print("\nCorrelation Matrix:")
corr_matrix = df.corr()
print(corr_matrix)

# Visualization: Correlation Heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()

# Step 2: Check for Multicollinearity
# Variance Inflation Factor (VIF) to detect multicollinearity
from statsmodels.stats.outliers_influence import variance_inflation_factor

# Calculate VIF for each feature
X = df[['Revenue', 'Gross Profit', 'EBITDA', 'Net Income', 'Total Assets', 'Total Liabilities']]
vif_data = pd.DataFrame()
vif_data["Feature"] = X.columns
vif_data["VIF"] = [variance_inflation_factor(X.values, i) for i in range(len(X.columns))]

print("\nVariance Inflation Factor (VIF):")
print(vif_data)

# Step 3: Residual Analysis (to assess model fit and detect anomalies)
# Perform a simple linear regression as an example
X = sm.add_constant(X)  # Adds a constant term to the model
model = sm.OLS(df['Stock Price'], X).fit()
predictions = model.predict(X)

# Plot residuals
residuals = df['Stock Price'] - predictions
plt.figure(figsize=(10, 6))
plt.scatter(predictions, residuals, color='purple')
plt.hlines(y=0, xmin=predictions.min(), xmax=predictions.max(), color='red')
plt.xlabel('Predicted Values')
plt.ylabel('Residuals')
plt.title('Residual Plot')
plt.show()

# Step 4: Regression Diagnostics (Assessing model assumptions)
# 1. Linearity: Check if relationships between predictors and target variable are linear
# 2. Normality: Check if the residuals are normally distributed
sm.qqplot(residuals, line='s')
plt.title('QQ Plot of Residuals')
plt.show()

# 3. Homoscedasticity: Check if residuals have constant variance
plt.figure(figsize=(10, 6))
plt.scatter(predictions, residuals, color='blue')
plt.hlines(y=0, xmin=predictions.min(), xmax=predictions.max(), color='red')
plt.xlabel('Predicted Values')
plt.ylabel('Residuals')
plt.title('Homoscedasticity Check (Residuals vs Predicted)')
plt.show()

# 4. Independence: Check for autocorrelation in residuals using Durbin-Watson test
from statsmodels.stats.stattools import durbin_watson
dw_test = durbin_watson(residuals)
print(f"\nDurbin-Watson Test Statistic: {dw_test}")

# Step 5: Outlier Detection (Detecting influential data points)
# Cook's Distance to detect influential data points
influence = model.get_influence()
cooks_d = influence.cooks_distance[0]

plt.figure(figsize=(10, 6))
plt.stem(cooks_d, markerfmt=",", use_line_collection=True)
plt.title("Cook's Distance")
plt.xlabel('Observation Index')
plt.ylabel("Cook's Distance")
plt.show()

# Identify points with high Cook's Distance
influential_points = df.index[cooks_d > 4 / len(df)]
print(f"\nInfluential Points (High Cook's Distance): {influential_points.tolist()}")

# Step 6: Interpretation and Explanation
# Summarize findings and interpret potential causes of observed patterns, anomalies, and relationships.

# Summary of the regression model
print("\nSummary of Regression Model:")
print(model.summary())
```
